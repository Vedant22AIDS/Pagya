```python
""" 
Group A: Assignment No.3
Assignment Title: Write a Python Program using Perceptron Neural
Network to recognise even and odd numbers. Given numbers are in ASCII
form 0 to 9
"""
import numpy as np

class ParityPerceptron:
    def __init__(self):
        self.weights = np.zeros(11)  # 10 digits + bias
        self.learning_rate = 0.1
    
    def predict(self, x):
        x = np.insert(x, 0, 1)  # Add bias term
        return 1 if np.dot(self.weights, x) >= 0 else 0
    
    def train(self, X, y):
        for _ in range(100):  # 100 epochs
            for xi, target in zip(X, y):
                prediction = self.predict(xi)
                error = target - prediction
                self.weights += self.learning_rate * error * np.insert(xi, 0, 1)

def main():
    # Training data (0-9 as one-hot vectors)
    X = np.eye(10, dtype=int)
    y = [i%2 for i in range(10)]  # 0=even, 1=odd
    
    # Create and train perceptron
    ppn = ParityPerceptron()
    ppn.train(X, y)
    
    # Test all numbers
    print("Number\tPrediction")
    for i in range(10):
        print(f"{i}\t{'Odd' if ppn.predict(X[i]) else 'Even'}")
    
    # Verify accuracy
    correct = sum(ppn.predict(x) == (i%2) for i, x in enumerate(X))
    print(f"\nAccuracy: {correct*10}%")

if __name__ == "__main__":
    main()
```

```python
# Import numpy for mathematical operations and arrays
import numpy as np

# Define a class for the Perceptron to recognize even and odd numbers
class ParityPerceptron:
    def __init__(self):
        # Initialize weights with zeros for 10 input digits + 1 bias term
        self.weights = np.zeros(11)
        # Set the learning rate for weight updates
        self.learning_rate = 0.1

    # Define prediction function
    def predict(self, x):
        # Add bias input (1) at the beginning of the input vector
        x = np.insert(x, 0, 1)
        # Calculate the dot product (weighted sum)
        # If sum >= 0, predict 1 (odd), else 0 (even)
        return 1 if np.dot(self.weights, x) >= 0 else 0

    # Define the training function
    def train(self, X, y):
        # Repeat for 100 iterations (epochs) to improve learning
        for _ in range(100):
            # Loop over each training example and its label
            for xi, target in zip(X, y):
                prediction = self.predict(xi)  # Get prediction
                error = target - prediction    # Calculate error
                # Update weights using perceptron learning rule
                self.weights += self.learning_rate * error * np.insert(xi, 0, 1)

# Main function to run training and testing
def main():
    # Create one-hot encoded inputs for digits 0 to 9
    X = np.eye(10, dtype=int)  # Each row represents a digit in ASCII (0â€“9)

    # Target output: 0 for even numbers, 1 for odd numbers
    y = [i % 2 for i in range(10)]

    # Create a perceptron object and train it with the data
    ppn = ParityPerceptron()
    ppn.train(X, y)

    # Test the perceptron on digits 0 to 9 and print predictions
    print("Number\tPrediction")
    for i in range(10):
        result = ppn.predict(X[i])
        print(f"{i}\t{'Odd' if result else 'Even'}")

    # Calculate accuracy of predictions
    correct = sum(ppn.predict(x) == (i % 2) for i, x in enumerate(X))
    print(f"\nAccuracy: {correct * 10}%")  # Since 10 samples, multiply by 10 for percentage

# Ensure the main function runs only if the script is executed directly
if __name__ == "__main__":
    main()
```
