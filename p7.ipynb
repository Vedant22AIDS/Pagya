```python
""" 
Group B: Assignment No. 1
Assignment Title: Write a python program to show Back Propagation
Network for XOR function with Binary Input and Output
"""

import numpy as np

# Sigmoid & Derivative
sigmoid = lambda x: 1 / (1 + np.exp(-x))
d_sigmoid = lambda x: x * (1 - x)

class XORNet:
    def __init__(self):
        self.W1 = np.random.randn(2, 2)
        self.b1 = np.random.randn(1, 2)
        self.W2 = np.random.randn(2, 1)
        self.b2 = np.random.randn(1, 1)

    def forward(self, X):
        self.h = sigmoid(X @ self.W1 + self.b1)
        self.o = sigmoid(self.h @ self.W2 + self.b2)
        return self.o

    def backward(self, X, y):
        err = y - self.o
        d_o = err * d_sigmoid(self.o)
        d_h = d_o @ self.W2.T * d_sigmoid(self.h)

        self.W2 += self.h.T @ d_o
        self.b2 += d_o.sum(0, keepdims=True)
        self.W1 += X.T @ d_h
        self.b1 += d_h.sum(0, keepdims=True)

    def train(self, X, y, epochs=10000):
        for _ in range(epochs):
            self.forward(X)
            self.backward(X, y)

    def predict(self, X):
        return self.forward(X)

# XOR Input and Output
X = np.array([[0,0], [0,1], [1,0], [1,1]])
y = np.array([[0], [1], [1], [0]])

# Train and Predict
model = XORNet()
model.train(X, y)
print("Predictions:\n", model.predict(X).round(2))
```

```python
""" 
Group B: Assignment No. 1
Assignment Title: Write a python program to show Back Propagation
Network for XOR function with Binary Input and Output
"""

import numpy as np

# Sigmoid activation function (adds non-linearity)
sigmoid = lambda x: 1 / (1 + np.exp(-x))

# Derivative of sigmoid (used to compute gradients in backpropagation)
d_sigmoid = lambda x: x * (1 - x)

# Define the XOR neural network
class XORNet:
    def __init__(self):
        # Initialize weights and biases with random values
        # 2 input neurons -> 2 hidden neurons
        self.W1 = np.random.randn(2, 2)   # Weights from input to hidden layer
        self.b1 = np.random.randn(1, 2)   # Biases for hidden layer

        # 2 hidden neurons -> 1 output neuron
        self.W2 = np.random.randn(2, 1)   # Weights from hidden to output layer
        self.b2 = np.random.randn(1, 1)   # Bias for output layer

    def forward(self, X):
        # Forward pass to compute predictions
        self.h = sigmoid(X @ self.W1 + self.b1)      # Hidden layer activation
        self.o = sigmoid(self.h @ self.W2 + self.b2) # Output layer activation
        return self.o

    def backward(self, X, y):
        # Backpropagation to adjust weights based on error

        err = y - self.o                    # Error at output
        d_o = err * d_sigmoid(self.o)      # Delta for output layer

        d_h = d_o @ self.W2.T * d_sigmoid(self.h)  # Delta for hidden layer

        # Update weights and biases
        self.W2 += self.h.T @ d_o           # Update weights from hidden to output
        self.b2 += d_o.sum(0, keepdims=True)  # Update output bias

        self.W1 += X.T @ d_h                # Update weights from input to hidden
        self.b1 += d_h.sum(0, keepdims=True)  # Update hidden bias

    def train(self, X, y, epochs=10000):
        # Train the network for a number of epochs
        for _ in range(epochs):
            self.forward(X)
            self.backward(X, y)

    def predict(self, X):
        # Predict output for given input
        return self.forward(X)

# Define input and expected output for XOR logic
X = np.array([[0,0], [0,1], [1,0], [1,1]])  # XOR inputs
y = np.array([[0], [1], [1], [0]])          # XOR expected outputs

# Create the model, train it, and print predictions
model = XORNet()
model.train(X, y)
print("Predictions:\n", model.predict(X).round(2))  # Rounded predictions for readability
```
